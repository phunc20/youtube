## The Macros and Their Meaning

| macro | meaning |
| ----- | ------- |
| `.TL` | title   |
| `.AU` | author   |
| `.AI` | institute   |
| `.NH` | numbered heading; section |
| `.SH` | unnumbered section |
| `.PP` | paragraph   |
| `.B`  | bold |
| `.I`  | italic |
| `.BI`  | bold and italic |
| `.UL`  | underline |
| `.IP \(bu 2`  | bullet list |




### Rules
- All macros has to be put at the front of a line and capitalized
- Macros like `.B` can actually accept three arguments
  01. The string to be rendered bold
  02. The string that comes **directly after** the bold part
  03. The string that comes **directly before** the bold part

```julia
content = """
          .TL
          My first groff document
          .AU
          Luke Smith
          .AI
          University of Arizona
          .NH
          First Section
          .PP
          This is a paragraph. Another
          sentence.
          And another.
          Yet another.
          
          Unindented paragraph here.
          .PP
          Indented paragraph has to be preceded by .P
          .NH
          Second Section
          .PP
          Some more text. I want the word
          .B "bold"
          to be bold. Want
          .B "small " pox
          to be bold as well.
          .NH 2
          Subsection
          .NH 1
          Conclusion
          .NH 2
          Subconclusion
          """;
```

```julia
open(f->write(f, content), "groff.ms", "w")
```

```julia
;cat groff.ms
```
